## 问题 3

给定一个包含n个元素的数组 `profits` , 它的第 `i` 个元素 `profits[i]` 表示一支股票第 `i` 天的**收益**（正数表示涨, 负数表示跌）。你只能选择 **某一天** 买入这只股票, 并选择在 **未来的某一个不同的日子** 卖出该股票。

1. 设计一个算法来计算你所能获取的最大利润和对应买入和卖出的日期。请分析算法方案, 计算其时间复杂度, 并且使用python编程实现该算法。

2. \* 设计一个时间复杂度为 $O(n)$的算法实现该算法

e.g. :
---
profits=[3,2,1,-7,5,2,-1,3,-1], 第5天买入, 第8天卖出, 收益最大：9

````python
时间复杂度：O(n)
````

````python
def max_profit(profits):
    """
    计算股票一次买卖的最大利润及对应日期
    :param profits: list[int], 每日收益
    :return: (max_profit, buy_day, sell_day)
             max_profit: 最大利润
             buy_day: 买入日期 (0-based 索引)
             sell_day: 卖出日期 (0-based 索引, 保证 > buy_day)
    """
    n = len(profits)
    if n == 0:
        return 0, -1, -1  # 无数据

    # 初始化：第一个元素作为起始子数组
    cur_sum = profits[0]
    max_sum = profits[0]
    start = 0       
    end = 0         
    temp_start = 0  
    for i in range(1, n):
        # 判断是否以当前元素开始新的子数组
        if cur_sum + profits[i] > profits[i]:
            cur_sum += profits[i]
        else:
            cur_sum = profits[i]
            temp_start = i

        # 更新全局最优
        if cur_sum > max_sum:
            max_sum = cur_sum
            start = temp_start
            end = i

    # 卖出日 = 子数组结束索引 + 1
    sell_day = end + 1
    return max_sum, start, sell_day


if __name__ == "__main__":
    # 测试用例1：一般情况
    profits1 = [3, 2, 1, -7, 5, 2, -1, 3, -1]
    profit, buy, sell = max_profit(profits1)
    print(f"最大利润: {profit}")
    print(f"买入日(索引): {buy}")
    print(f"卖出日(索引): {sell}")
````

## 问题 4

给定一个包含不同整数元素的数组 $ A[1..n]$ ,并且满足条件：$A[1]>A[2]$ 并且 $A[n-1]<A[n]$; 规定：如果一个元素比它两边的邻居元素都小, 即：$A[x]<A[x-1], A[x]<A[x+1]$ , 称这个元素A[x]为“局部最小”。通过遍历一次数组, 我们可以很容易在 $O(n)$的时间复杂度下找到一个局部最小值, 


1. 分析该问题, 设计一个算法在$O(logn)$的时间复杂度下找到一个局部最小(返回数值), 要求：分析算法设计思路, 并且使用python编程实现
2. \* 设计算法找出所有局部最小值, 分析算法设计思路, 并使用python编程实现

e.g.:
---
A=[9, 3, 7, 2, 1, 4, 5 ] 时,  局部最小元素为 3, 1

idea:时间复杂度 O(log n)

````python
def local_minimum(arr):
    """
    在满足 A[0] > A[1] 且 A[-2] < A[-1] 的互异整数数组中，
    以 O(log n) 时间返回任意一个局部最小值的数值。
    """
    n = len(arr)
    # 保证题目条件成立，n >= 3
    left, right = 0, n - 1

    while left <= right:
        mid = (left + right) // 2

        # 检查 mid 是否为局部最小
        # 需要确保邻居存在
        left_ok = (mid == 0) or (arr[mid] < arr[mid-1])
        right_ok = (mid == n-1) or (arr[mid] < arr[mid+1])

        if left_ok and right_ok:
            return arr[mid]

        # 否则，向值更小的一侧收缩
        # 由于 arr[mid] 不是局部最小，至少有一侧邻居比它小
        if mid > 0 and arr[mid-1] < arr[mid]:
            right = mid - 1
        else:
            left = mid + 1

    # 理论上总会找到，但若未找到，返回最小值的索引（备用）
    # 实际上 while 内部一定会返回
    return min(arr)  # fallback

# 测试用例
if __name__ == "__main__":
    A = [9, 3, 7, 2, 1, 4, 5]  
    print(local_minimum(A)) 
````

e.g.：

数组arr=[5,4,3,2,6,1,88,33,22,107] 的前3个最小数据为：1, 2, 3

````python
def sort_and_get_k_smallest(arr, k):
    """
    先对数组排序，再返回最小的 k 个数
    :param arr: list, 包含 n 个不同数字的数组
    :param k: int, 需要返回的最小元素个数
    :return: list, 按升序排列的最小 k 个数
    """
    if k <= 0:
        return []
    # 使用内置排序（Timsort，O(n log n)）
    sorted_arr = sorted(arr)
    return sorted_arr[:k]

# 测试
if __name__ == "__main__":
    arr=[5,4,3,2,6,1,88,33,22,107] 
    k = 3
    result = sort_and_get_k_smallest(arr, k)
    print(f"最小的{k}个数: {result}")   # 输出: 最小的3个数: [1, 2, 3]
````

````python
import heapq

def heap_method_get_k_smallest(arr, k):
    """
    使用空堆逐个插入建堆，然后 k 次提取最小值，返回最小的 k 个数
    :param arr: list, 包含 n 个不同数字的数组
    :param k: int, 需要返回的最小元素个数
    :return: list, 按升序排列的最小 k 个数
    """
    if k <= 0:
        return []
    # 建空堆
    heap = []
    # 逐个插入
    for val in arr:
        heapq.heappush(heap, val)   # O(log size)
    # 提取 k 次最小值
    result = []
    for _ in range(min(k, len(heap))):
        result.append(heapq.heappop(heap))  # O(log size)
    return result

# 测试
if __name__ == "__main__":
    arr=[5,4,3,2,6,1,88,33,22,107] 
    k = 3
    result = heap_method_get_k_smallest(arr, k)
    print(f"最小的{k}个数 (堆方法): {result}")   # 输出: [1, 2, 3]
````

## 问题 6

> 星球碰撞问题：现有n个星球，在同一条直线上运行，如数组A所示，元素的绝对值表示星球的质量，负数表示星球自右向左运动，正数表示星球自左向右运动，当两个星球相撞的时候，质量小的会消失，大的保持不变，**质量相同的两个星球碰撞后自右向左运动的星球消失，自左向右的星球保持不变**，假设所有星球的速度大小相同。
>
> $ A=[23,-8, 9, -3, -7, 9, -23, 22] $

请设计一个算法模拟星球的运行情况，输出最终的星球存续情况（输出一个数组），分析算法设计思路，计算时间复杂度，并基于python编程实现。

e.g.
---
1.  输入： A=[-3,-6,2,8, 5,-8,9,-2,1]， 输出：[-3, -6, 2, 8, 9, 1]

2. 输入：A=[23,-8, 9, -3, -7, 9, -23, 22], 输出：[23, 22]

idea:时间复杂度：O(n)

````python
def asteroid_collision(asteroids):
    """
    模拟星球碰撞过程
    :param asteroids: List[int], 每个元素表示星球的质量和方向（正→右，负→左）
    :return: List[int], 碰撞后存活的星球，保持原始顺序
    """
    stack = []
    for a in asteroids:
        while stack and stack[-1] > 0 and a < 0:
            top = stack[-1]
            if abs(top) > abs(a):
                break
            elif abs(top) < abs(a):
                stack.pop()
                continue
            else: 
                break
        else:
            # 没有发生碰撞或碰撞结束后当前星球存活，则入栈
            stack.append(a)
    return stack


if __name__ == "__main__":
    # 示例 1
    A1 = [-3, -6, 2, 8, 5, -8, 9, -2, 1]
    print(asteroid_collision(A1))  # 输出: [-3, -6, 2, 8, 9, 1]

    # 示例 2
    A2 = [23, -8, 9, -3, -7, 9, -23, 22]
    print(asteroid_collision(A2))  
````

## 问题 7

> 0-1背包问题：现在有4块大理石，每个大理石对应的重量和价值使用一个元组表示, 即（重量，价值），4块大理石的重量和价值为：[(5,10), (4,40), (6,30), (3, 50)], 假设你有一辆最大承重为9的小推车，请问使用这个小推车装大理石的最大价值为多少

请设计一个算法解决该问题，分析算法设计思路，计算时间复杂度，并基于python编程实现

idea:时间复杂度为o(n)

````python
def knapsack_01(weights, values, capacity):
    """
    0-1背包问题，返回最大价值
    :param weights: list[int], 每个物品的重量
    :param values:  list[int], 每个物品的价值
    :param capacity: int, 背包容量
    :return: int, 最大价值
    """
    n = len(weights)
    # 一维DP数组，初始化为0
    dp = [0] * (capacity + 1)

    for i in range(n):
        # 倒序遍历容量，保证每个物品只取一次
        for j in range(capacity, weights[i] - 1, -1):
            dp[j] = max(dp[j], dp[j - weights[i]] + values[i])
    return dp[capacity]


if __name__ == "__main__":
    # 输入数据：每个物品为(重量, 价值)
    items = [(5, 10), (4, 40), (6, 30), (3, 50)]
    weights = [w for w, v in items]
    values = [v for w, v in items]
    capacity = 9

    max_value = knapsack_01(weights, values, capacity)
    print(f"最大价值为: {max_value}")  # 输出: 90
````

e.g.:
---
1.

输入：root = [1,3,null,null,2]， 输出为：[3,1,null,null,2]， 

解释：3 不能是 1 的左孩子，因为 3 > 1 。交换 1 和 3 使二叉搜索树有效。

2.

 <div style="text-align:center;width:200px;"> <img alt="" src="./fig/hw3q9-1.jpg"> </div>
输入： root=[3,1,4,null,null,2]， 输出为：[2,1,4,null,null,3]

解释：2 不能在 3 的右子树中，因为 2 < 3 。交换 2 和 3 使二叉搜索树有效。

idea:时间复杂度： O(n)

````python
# 定义二叉树节点
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def recoverTree(root):
    """
    恢复二叉搜索树（两个节点被错误交换）
    :param root: TreeNode, 树的根节点
    :return: None, 直接修改原树
    """
    # 中序遍历收集节点
    nodes = []
    def inorder(node):
        if not node:
            return
        inorder(node.left)
        nodes.append(node)
        inorder(node.right)
    
    inorder(root)
    
    # 找到两个错误的节点
    first = second = None
    for i in range(len(nodes) - 1):
        if nodes[i].val > nodes[i+1].val:
            if first is None:
                first = nodes[i]
            second = nodes[i+1]  # 每次都更新，当第二次逆序时second就是第二个节点
    # 交换值
    if first and second:
        first.val, second.val = second.val, first.val

# 辅助函数：根据层次遍历列表构建树（用于测试）
def build_tree_from_level_order(lst):
    if not lst:
        return None
    from collections import deque
    root = TreeNode(lst[0])
    queue = deque([root])
    i = 1
    while queue and i < len(lst):
        node = queue.popleft()
        if i < len(lst) and lst[i] is not None:
            node.left = TreeNode(lst[i])
            queue.append(node.left)
        i += 1
        if i < len(lst) and lst[i] is not None:
            node.right = TreeNode(lst[i])
            queue.append(node.right)
        i += 1
    return root

# 辅助函数：层次遍历打印树（验证结果）
def level_order_print(root):
    if not root:
        return []
    from collections import deque
    res = []
    queue = deque([root])
    while queue:
        node = queue.popleft()
        if node:
            res.append(node.val)
            queue.append(node.left)
            queue.append(node.right)
        else:
            res.append(None)
    # 去除末尾的 None
    while res and res[-1] is None:
        res.pop()
    return res

if __name__ == "__main__":
    # 示例1: 输入 [1,3,null,null,2] 构建树
    # 根据LeetCode 99题，这个输入表示：
    #     1
    #      \
    #       3
    #      /
    #     2
    # 中序遍历应为 [1,2,3] 但实际为 [1,3,2]，交换3和2即可恢复。
    root1 = build_tree_from_level_order([1,3,None,None,2])
    recoverTree(root1)
    print(level_order_print(root1))  # 输出应接近 [3,1,null,null,2] 但具体层次结构不变
    
    # 示例2: 输入 [3,1,4,null,null,2]
    # 树结构:
    #        3
    #       / \
    #      1   4
    #         /
    #        2
    # 中序遍历应为 [1,2,3,4] 实际为 [1,3,2,4]，交换3和2即可。
    root2 = build_tree_from_level_order([3,1,4,None,None,2])
    recoverTree(root2)
    print(level_order_print(root2))  # 输出应接近 [2,1,4,null,null,3]
````

## 问题 10

给你一个二叉树的根节点 root ，树中每个节点都存放有一个 0 到 9 之间的数字。每条从根节点到叶节点的路径都代表一个数字：

> 例如，从根节点到叶节点的路径 1 -> 2 -> 3 表示数字 123 。

设计一个算法计算从根节点到叶节点生成的所有数字之和 ，分析算法设计思路，计算时间复杂度，并基于python编程实现

叶节点:是指没有子节点的节点。

e.g.:
---
1.

 <div style="text-align:center;width:200px;"> <img alt="figure" src="./fig/hw3q10-1.jpg"> </div>

 输入：root = [1,2,3], 输出：25

解释：从根到叶子节点路径 1->2 代表数字 12，1->3 代表数字 13，因此和为 12 + 13 = 25 。

2.

 <div style="text-align:center;width:200px;"> <img alt="figure" src="./fig/hw3q10-2.jpg"> </div>

 输入：root = [4,9,0,5,1], 输出：1026

解释：从根到叶子节点路径 4->9->5 代表数字 495，4->9->1 代表数字 491，4->0 代表数字 40。因此和为 495 + 491 + 40 = 1026 。

idea:时间复杂度：O(n)

````python
# 定义二叉树节点
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def sumNumbers(root):
    """
    计算所有根到叶子节点路径数字之和
    :param root: TreeNode, 树的根节点
    :return: int, 总和
    """
    def dfs(node, current_sum):
        if not node:
            return 0
        # 更新当前路径数字
        current_sum = current_sum * 10 + node.val
        # 如果是叶子节点，返回该路径数字
        if not node.left and not node.right:
            return current_sum
        # 否则递归左右子树并返回和
        return dfs(node.left, current_sum) + dfs(node.right, current_sum)
    
    return dfs(root, 0)


# 辅助函数：根据层次遍历列表构建树（用于测试）
def build_tree_from_level_order(lst):
    if not lst:
        return None
    from collections import deque
    root = TreeNode(lst[0])
    queue = deque([root])
    i = 1
    while queue and i < len(lst):
        node = queue.popleft()
        if i < len(lst) and lst[i] is not None:
            node.left = TreeNode(lst[i])
            queue.append(node.left)
        i += 1
        if i < len(lst) and lst[i] is not None:
            node.right = TreeNode(lst[i])
            queue.append(node.right)
        i += 1
    return root


if __name__ == "__main__":
    # 示例1: root = [1,2,3]
    root1 = build_tree_from_level_order([1, 2, 3])
    print(sumNumbers(root1))  # 输出: 25

    # 示例2: root = [4,9,0,5,1]
    root2 = build_tree_from_level_order([4, 9, 0, 5, 1])
    print(sumNumbers(root2))  # 输出: 1026

    # 附加测试: 单节点树
    root3 = TreeNode(5)
    print(sumNumbers(root3))  # 输出: 5
````
