## B 长跑

```python
#include <bits/stdc++.h>
using namespace std;

// 使用 long long 防止运算溢出
typedef long long ll;

int main() {
    // 优化输入输出效率
    ios::sync_with_stdio(false);
    cin.tie(0);

    ll n, l, maxn, s;
    while (cin >> n >> l >> maxn >> s) {
        // 使用 map 来存储站点：key是位置，value是该位置的最小花费
        // 这样可以自动处理同一位置多个站点的问题，且自动按位置排序
        map<ll, int> supply;

        for (int i = 0; i < n; i++) {
            ll p;
            int c;
            cin >> p >> c;
            // 只有在终点之前的站点才有意义（终点处的处理在后面统一覆盖）
            if (p < l) { 
                if (supply.find(p) == supply.end() || c < supply[p]) {
                    supply[p] = c;
                }
            }
        }

        // 关键步骤 1：强制设置起点（位置0）花费为0
        // 含义：初始满体力 = 在位置0免费加了一次油
        supply[0] = 0;

        // 关键步骤 2：强制设置终点（位置L）花费为0
        // 含义：到达终点不需要加油，费用为0用于占位，方便DP计算到达终点的代价
        supply[l] = 0;

        // 将 map 转换为 vector 以便通过索引访问
        vector<pair<ll, int>> stations;
        for (auto const& [pos, cost] : supply) {
            stations.push_back({pos, cost});
        }
        /*
        vector<pair<ll, int>> stations;
        for (auto const& item : supply) {
            ll position = item.first;
            int price = item.second;
            stations.push_back({position, price});
        }
        */

        int m = stations.size();
        // dp[i] 表示到达第 i 个站点（并补满/到达）的最小花费
        // 初始化为无穷大
        vector<ll> dp(m, 2e18);

        // 起点是 map 中的第一个元素（因为 map 自动按 key 排序，且 key=0 是最小的）
        dp[0] = 0;

        // O(N^2) 动态规划
        for (int i = 0; i < m; i++) {
            // 如果当前站点不可达，跳过
            if (dp[i] > s) continue;

            // 尝试从站点 i 跳到站点 j
            for (int j = i + 1; j < m; j++) {
                ll dist = stations[j].first - stations[i].first;

                // 如果距离超过体力上限，因为是排好序的，后面的更到不了，直接 break
                if (dist > maxn) {
                    break;
                }

                // 计算到达 j 的花费：到达 i 的花费 + j 处的加油费
                // 注意：如果 j 是终点 L，stations[j].second 是 0，费用不会增加，符合逻辑
                ll new_cost = dp[i] + stations[j].second;

                // 更新最小值
                if (new_cost < dp[j]) {
                    dp[j] = new_cost;
                }
            }
        }

        // map 自动排序，终点 L 一定是最后一个元素
        if (dp[m - 1] <= s) {
            cout << "Yes" << endl;
        } else {
            cout << "No" << endl;
        }
    }
    return 0;
}
```

## C 最长回文

```python
#include<bits/stdc++.h>
#define int long long
#define ull unsigned long long
using namespace std;
const int N = 3e5 + 10;
const int MOD = 999998639; 
const int P = 13331;
string a , b;
int pa[N] , pb[N] , res = 1;
ull pre[N] , suf[N] , power[N];
string Manacher(string a , int *p)
{
    string t = "$#";
    for(auto i : a) t += i , t += '#';
    int mx = 0 , id = 0 ;
    int len = t.size() , ans = 0; 
    for(int i = 1 ; i < len ; i ++)
    {
        p[i] = mx > i ? min(p[2 * id - i] , mx - i) : 1;
        while(t[i + p[i]] == t[i - p[i]]) p[i] ++ ;
        if(mx < i + p[i]) mx = i + p[i] , id = i;
        ans = max(ans , p[i] - 1);
    }
    res = max(res , ans);
    return t;
}
ull get_hash1(int l , int r)
{
    if(l > r) return -999;
    return (pre[r] - pre[l - 1] * power[r - l + 1] % MOD + MOD) % MOD;
}
ull get_hash2(int l , int r)
{
    if(l > r) return -888;
    return (suf[l] - suf[r + 1] * power[r - l + 1] % MOD + MOD) % MOD;
}
void init(int n)
{
    power[0] = 1;
    for(int i = 1 ; i < N - 5 ; i ++) power[i] = power[i - 1] * P % MOD;
    pre[0] = a[0] ;
    for(int i = 1 ; i < n ; i ++) pre[i] = (pre[i - 1] * P + a[i]) % MOD ;
    suf[n - 1] = b[n - 1];
    for(int i = n - 2 ; i >= 0 ; i --) suf[i] = (suf[i + 1] * P + b[i]) % MOD; 
}
signed main()
{
    int n ;
    cin >> n >> a >> b;
    a = Manacher(a , pa) , b = Manacher(b , pb);
    n = n * 2 + 2;
    init(n);
    int ans = 1;
    for(int i = 2 ; i <= n ; i ++)
    {
        int add = 0;
        int len = max(pa[i] , pb[i - 2]);
        int sa = i - len + 1  , sb = i - 2 + len - 1;
        int l = 0 , r = min(sa - 1 , n - sb);
        while(l <= r)
        {
            int mid = l + r >> 1;
            int l1 = max(0LL , sa - mid) , r1 = min(n , sa - 1);
            int l2 = max(sb + 1 , 0LL) , r2 = min(n , sb + mid);
            if(get_hash1(l1 , r1) == get_hash2(l2 , r2)) 
            l = mid + 1 , add = mid;
            else r = mid - 1; 
        }
        ans = max(ans , add + len - 1);
    }
    cout << ans << '\n';
    return 0;
}
```

## D 优惠券

```python
#include <iostream>
#include <vector>
#include <string>
#include <algorithm>
#include <numeric>


const int MAXM = 500005;
const int MAXX = 100005;

struct Record {
    char type;
    int id;
    
    Record() : type('?'), id(0) {}
    Record(char t, int i) : type(t), id(i) {}
};

std::vector<Record> recs;
std::vector<int> state_v(MAXX, 0);    
std::vector<int> last_pos(MAXX, 0);   
std::vector<int> q_pos;               
std::vector<int> fa;                  


int find_fa(int x) {
    if (fa[x] != x) {
        fa[x] = find_fa(fa[x]);
    }
    return fa[x];
}

/*
    尝试使用一个位置在 (L, R) 中的 ?
*/
bool use_question(int L, int R) {
    // 使用 C++ 标准库的 lower_bound 代替手写二分
    // 寻找 q_pos 中第一个大于 L 的元素的迭代器
    auto it = std::upper_bound(q_pos.begin(), q_pos.end(), L);
    
    // 转换为原本基于 1 开始的索引编号
    int idx = std::distance(q_pos.begin(), it) + 1;
    
    idx = find_fa(idx);
    int q_cnt = q_pos.size();

    if (idx <= q_cnt && q_pos[idx - 1] < R) {
        fa[idx] = find_fa(idx + 1);
        return true;
    }

    return false;
}

int solve_one_case() {
    int m;
    if (!(std::cin >> m)) {
        return -2;
    }

    recs.assign(m + 1, Record());
    q_pos.clear();
    std::fill(state_v.begin(), state_v.end(), 0);
    std::fill(last_pos.begin(), last_pos.end(), 0);

    for (int i = 1; i <= m; i++) {
        std::string op;
        std::cin >> op;
        recs[i].type = op[0];

        if (op[0] == 'I' || op[0] == 'O') {
            std::cin >> recs[i].id;
        } else {
            recs[i].id = 0;
            q_pos.push_back(i); // 收集 '?' 的位置
        }
    }

    int q_cnt = q_pos.size();
    fa.resize(q_cnt + 2);
    // 并查集初始化：fa[i] = i
    std::iota(fa.begin(), fa.end(), 0);

    for (int i = 1; i <= m; i++) {
        char op = recs[i].type;

        if (op == '?') {
            continue;
        }

        int x = recs[i].id;

        if (op == 'I') {
            if (state_v[x] == 0) {
                state_v[x] = 1;
                last_pos[x] = i;
            } else {
                if (!use_question(last_pos[x], i)) {
                    return i;
                }
                state_v[x] = 1;
                last_pos[x] = i;
            }
        } else if (op == 'O') {
            if (state_v[x] == 1) {
                state_v[x] = 0;
                last_pos[x] = i;
            } else {
                if (!use_question(last_pos[x], i)) {
                    return i;
                }
                state_v[x] = 0;
                last_pos[x] = i;
            }
        }
    }

    return -1;
}

int main() {
    std::ios_base::sync_with_stdio(false);
    std::cin.tie(NULL);

    while (true) {
        int ans = solve_one_case();

        if (ans == -2) {
            break;
        }

        std::cout << ans << "\n";
    }

    return 0;
}
```

## F 通配符匹配

```python
## add your code here#include <iostream>
#include <vector>
#include <string>
#include <algorithm>
#include <cstring>

// 保持原本的常数限制
const int MAXITEM = 105;

struct Segment {
    int len;                    // 当前段长度
    int cnt;                    // 当前段中的普通字符块数量
    std::vector<int> off;       // 普通字符块在当前段中的偏移
    std::vector<int> cid;       // 普通字符块编号
    int anchor;                 // 最长普通字符块，用于快速定位

    Segment() : len(0), cnt(0), anchor(-1) {
        off.resize(MAXITEM, 0);
        cid.resize(MAXITEM, 0);
    }
};

struct Chunk {
    int start;                  // 普通字符块在模式串中的起始位置
    int len;                    // 普通字符块长度
    std::vector<int> pi;        // KMP 前缀数组
};

// 使用全局或静态容器存储，避免每次调用重新分配影响性能
std::string pattern, text;
int plen;
int segCnt, chunkCnt, starCnt;

std::vector<Segment> segs(MAXITEM);
std::vector<Chunk> chunks(MAXITEM);

// 动态分配匹配矩阵以优化空间占用
std::vector<std::vector<unsigned char>> occ;
std::vector<std::vector<int>> posList;
std::vector<int> posCnt;

void build_pi(int id) {
    int len = chunks[id].len;
    int start = chunks[id].start;

    chunks[id].pi.assign(len, 0);
    auto& pi = chunks[id].pi;

    for (int i = 1; i < len; i++) {
        int j = pi[i - 1];

        while (j > 0 && pattern[start + i] != pattern[start + j]) {
            j = pi[j - 1];
        }

        if (pattern[start + i] == pattern[start + j]) {
            j++;
        }

        pi[i] = j;
    }
}

void parse_pattern() {
    plen = pattern.length();

    segCnt = 0;
    chunkCnt = 0;
    starCnt = 0;

    for (int i = 0; i < plen; i++) {
        if (pattern[i] == '*') {
            starCnt++;
        }
    }

    int segStart = 0;

    while (true) {
        int segEnd = segStart;

        while (segEnd < plen && pattern[segEnd] != '*') {
            segEnd++;
        }

        Segment *sg = &segs[segCnt];
        sg->len = segEnd - segStart;
        sg->cnt = 0;
        sg->anchor = -1;

        int j = segStart;

        while (j < segEnd) {
            while (j < segEnd && pattern[j] == '?') {
                j++;
            }

            int st = j;

            while (j < segEnd && pattern[j] != '?') {
                j++;
            }

            if (j > st) {
                int id = chunkCnt++;

                chunks[id].start = st;
                chunks[id].len = j - st;

                build_pi(id);

                int c = sg->cnt++;
                sg->off[c] = st - segStart;
                sg->cid[c] = id;

                if (sg->anchor == -1 ||
                    chunks[id].len > chunks[sg->cid[sg->anchor]].len) {
                    sg->anchor = c;
                }
            }
        }

        segCnt++;

        if (segEnd == plen) {
            break;
        }

        segStart = segEnd + 1;
    }
}

void build_occurrences(int n) {
    // 动态调整表格尺寸，省去多余内存清空时间
    occ.assign(chunkCnt, std::vector<unsigned char>(n + 1, 0));
    posList.assign(chunkCnt, std::vector<int>(n + 1, 0));
    posCnt.assign(chunkCnt, 0);

    for (int id = 0; id < chunkCnt; id++) {
        int m = chunks[id].len;

        if (m > n) {
            continue;
        }

        int start = chunks[id].start;
        const auto& pi = chunks[id].pi;

        int j = 0;

        for (int i = 0; i < n; i++) {
            while (j > 0 && text[i] != pattern[start + j]) {
                j = pi[j - 1];
            }

            if (text[i] == pattern[start + j]) {
                j++;
            }

            if (j == m) {
                int pos = i - m + 1;

                occ[id][pos] = 1;
                posList[id][posCnt[id]++] = pos;

                j = pi[j - 1];
            }
        }
    }
}

int segment_at(Segment *sg, int n, int st) {
    if (st < 0 || st + sg->len > n) {
        return 0;
    }

    for (int i = 0; i < sg->cnt; i++) {
        int id = sg->cid[i];
        int pos = st + sg->off[i];

        if (!occ[id][pos]) {
            return 0;
        }
    }

    return 1;
}

int find_segment(Segment *sg, int n, int pos) {
    if (pos + sg->len > n) {
        return -1;
    }

    // 如果这一段没有普通字符，说明这一段全是 '?'
    if (sg->cnt == 0) {
        return pos;
    }

    int a = sg->anchor;
    int id = sg->cid[a];

    int need = pos + sg->off[a];
    
    // 使用 std::lower_bound 替代手写二分查找
    auto begin_it = posList[id].begin();
    auto end_it = begin_it + posCnt[id];
    auto it = std::lower_bound(begin_it, end_it, need);
    int idx = std::distance(begin_it, it);

    while (idx < posCnt[id]) {
        int st = posList[id][idx] - sg->off[a];

        if (st + sg->len > n) {
            break;
        }

        if (st >= pos && segment_at(sg, n, st)) {
            return st;
        }

        idx++;
    }

    return -1;
}

int wildcard_match() {
    int n = text.length();

    build_occurrences(n);

    // 如果没有 '*'，必须整体完全匹配
    if (starCnt == 0) {
        return segs[0].len == n && segment_at(&segs[0], n, 0);
    }

    int pos = 0;
    int first = 0;

    // 如果模式串开头不是 '*'，第一段必须从文件名开头匹配
    if (pattern[0] != '*') {
        if (!segment_at(&segs[0], n, 0)) {
            return 0;
        }

        pos = segs[0].len;
        first = 1;
    }

    // 如果模式串结尾不是 '*'，最后一段必须贴着文件名结尾匹配
    bool hasTail = pattern[plen - 1] != '*';
    int endSeg = segCnt - (hasTail ? 1 : 0);

    for (int i = first; i < endSeg; i++) {
        int st = find_segment(&segs[i], n, pos);

        if (st == -1) {
            return 0;
        }

        pos = st + segs[i].len;
    }

    if (hasTail) {
        Segment *tail = &segs[segCnt - 1];
        int st = n - tail->len;

        if (st < pos) {
            return 0;
        }

        if (!segment_at(tail, n, st)) {
            return 0;
        }
    }

    return 1;
}

int main() {
    // 提升 cin/cout 性能，使其超越 scanf/printf
    std::ios_base::sync_with_stdio(false);
    std::cin.tie(NULL);

    int n;
    if (std::cin >> pattern >> n) {
        parse_pattern();

        for (int i = 0; i < n; i++) {
            std::cin >> text;

            if (wildcard_match()) {
                std::cout << "YES\n";
            } else {
                std::cout << "NO\n";
            }
        }
    }

    return 0;
}
```

## H 马步距离

```python
#include<stdio.h>
#include<iostream>
using namespace std;
const int N=10005,G[10][2]={{1,2},{1,-2},{-1,2},{-1,-2},{2,1},{2,-1},{-2,1},{-2,-1}};
struct Point{
	int x,y;
};
struct queue{
	int l,r;
	Point a[N];
	void clear(){
		l=0,r=1;
	}
	bool empty(){
		return l+1==r;
	}
	void push(Point x){
		a[r]=x;
		r=(r+1)%N;
	}
	void pop(){
		l=(l+1)%N;
	}
	Point front(){
		return a[(l+1)%N];
	}
}q;
bool vis[105][105];
int x,y,d[105][105];
int abs(int n){
	return n>0?n:-n;
}
int BFS(){
	q.clear();
	q.push((Point){x,y});
	vis[x][y]=1;
	Point p;
	int i;
	while(!vis[50][50]){
		p=q.front();
		q.pop();
		for(i=0;i<8;i++){
			if(p.x+G[i][0]>=0&&p.x+G[i][0]<=100&&p.y+G[i][1]>=0&&p.y+G[i][1]<=100&&!vis[p.x+G[i][0]][p.y+G[i][1]]){
				vis[p.x+G[i][0]][p.y+G[i][1]]=1;
				d[p.x+G[i][0]][p.y+G[i][1]]=d[p.x][p.y]+1;
				q.push((Point){p.x+G[i][0],p.y+G[i][1]});
			}
		}
	}
	return d[50][50];
}
int main(){
	int a,b,s=0;
	scanf("%d%d%d%d",&x,&y,&a,&b);
	x=abs(x-a);
	y=abs(y-b);
	while(x+y>=50){
		if(x<y){
			swap(x,y);
		}
		x-=4;
		if(x-4<=y*2){
			y-=2;
		}
		s+=2;
	}
	x+=50;
	y+=50;
	printf("%d\n",s+BFS());
	return 0;
}
```

## I 直方图最大矩形

```python
class Solution {
public:
    /**
     * 代码中的类名、方法名、参数名已经指定，请勿修改，直接返回方法规定的值即可
     *
     * @param heights int整型vector 
     * @return int整型
     */
    int largestRectangleArea(vector<int>& heights) {
        int max_area = 0;
        // 创建一个辅助常规数组，前后加 0 作为哨兵
        vector<int> tmp;
        tmp.push_back(0);
        for (int h : heights) {
            tmp.push_back(h);
        }
        tmp.push_back(0);

        // 单调栈，存储下标
        vector<int> st;
        
        for (int i = 0; i < tmp.size(); ++i) {
            // 当当前高度小于栈顶高度时，说明栈顶元素的右边界已找到，触发计算
            while (!st.empty() && tmp[i] < tmp[st.back()]) {
                int cur_height = tmp[st.back()];
                st.pop_back(); // 弹出当前要计算高度的柱子
                
                // 弹出后，新的栈顶就是左边界
                int left_boundary = st.back();
                int width = i - left_boundary - 1;
                
                max_area = max(max_area, cur_height * width);
            }
            // 当前元素入栈，继续保持单调递增
            st.push_back(i);
        }

        return max_area;
    }
};
```

## J 消防局的设立

```python
#include<bits/stdc++.h>
#define ll long long
using namespace std;
const ll INF=1e18;
ll n,y,dp[1005][5];
vector<ll>t[100005];
void dfs(ll x,ll fa){
	ll cnt=0,s2=0,s3=0;
	for(auto y:t[x]){
		if(y==fa) continue;
		dfs(y,x);
		s2+=dp[y][2];
		s3+=dp[y][3];
		cnt++;
	}
	if(!cnt){
		dp[x][0]=dp[x][1]=dp[x][2]=1;
		dp[x][3]=dp[x][4]=0;
		return;
	}
	dp[x][0]=1;dp[x][1]=dp[x][2]=INF;
	dp[x][3]=0;dp[x][4]=0;
	for(auto y:t[x]){
		if(y==fa) continue;
		dp[x][0]+=dp[y][4];
		dp[x][1]=min(dp[y][0]+s3-dp[y][3],dp[x][1]);
		dp[x][2]=min(dp[y][1]+s2-dp[y][2],dp[x][2]);
		dp[x][3]+=dp[y][2];
		dp[x][4]+=dp[y][3];
	}
	for(int i=1;i<5;i++)
		dp[x][i]=min(dp[x][i],dp[x][i-1]);
}
int main(){
	cin>>n;
	for(int i=1;i<n;i++){
		cin>>y;
		t[i+1].push_back(y);
		t[y].push_back(i+1);
	}
	dfs(1,0);
	cout<<dp[1][2];
	return 0;
}
```
